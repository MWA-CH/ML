# PyPlot vs Object Oriented Method

```python
heart = pd.read_csv('heart.csv.xls')
heart
```

```python
heart.head(10)
```

```python
over_40 = heart[heart['age'] > 40]
over_40
```

```python
# pyplot method
over_40.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
# object oriented approch
fig,ax = plt.subplots(figsize=(6,5))
over_40.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
```

# Entire Life Cycle of Object Oriented Plot

```python
# create figure and graph area
fig,ax = plt.subplots(figsize=(10,6))

#plot the data
scatter_plot = ax.scatter(x=over_40['age'],y=over_40['chol'],c=over_40['target'])

#customization
ax.set(title='Suspectability of Heart Diesease',xlabel='Age',ylabel='Cholestorol')

#set the legend
ax.legend(*scatter_plot.legend_elements(), title='target')

scatter_plot;
```

# Entire Life Cycle of Object Oriented Plot Advance

```python
#1 create plot area
fig,(plot1,plot2) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

#2 Add data to plot
scatter_plot = plot1.scatter(x=over_40['age'],y=over_40['chol'],c=over_40['target'],cmap='winter',)

#3 customise
plot1.set(title='Heart Disease Problem',xlabel='Age',ylabel='Cholestorol')

#4 Add legend
plot1.legend(*scatter_plot.legend_elements(),title='Servived or Not')

#5 Meanline
plot1.axhline(y=over_40['chol'].mean(),linestyle="--");

#now plot second plot
scatter_plot = plot2.scatter(x=over_40['age'],y=over_40['trestbps'],c=over_40['target'],cmap='plasma')

#customise
plot2.set(title='Heart Disease Problem with Resting Blood Pressure',xlabel='Age',ylabel='Resting Blood Pressure')

#Add legend
plot2.legend(*scatter_plot.legend_elements(),title='Servived or Not');

#Meanline
plot2.axhline(y=over_40['trestbps'].mean(),linestyle="--");

#6 Coloring Data

#7 Coloring Plot
plt.style.use('seaborn-v0_8-whitegrid')

#8 Naming the figure
fig.suptitle('Patients Data',fontsize=32,fontweight='bold'); #tab + shift to see method attributes

#9 Saving the figure
fig.savefig('Patients data report.png');
```
